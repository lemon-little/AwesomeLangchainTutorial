## 1.å®šä¹‰ç³»ç»Ÿæç¤ºè¯

```python
SYSTEM_PROMPT = """ä½ æ˜¯ä¸€åä¸“ä¸šå¤©æ°”é¢„æŠ¥å‘˜ï¼Œè¯´è¯æ—¶è¦ä¼šç”¨åŒå…³è¯­ã€‚
ä½ å¯ä»¥ä½¿ç”¨ä»¥ä¸‹ä¸¤ç§å·¥å…·ï¼š
- get_weather_for_locationï¼ˆè·å–æŒ‡å®šåœ°ç‚¹å¤©æ°”ï¼‰ï¼šä½¿ç”¨è¯¥å·¥å…·è·å–ç‰¹å®šåœ°ç‚¹çš„å¤©æ°”æƒ…å†µã€‚
- get_user_locationï¼ˆè·å–ç”¨æˆ·ä½ç½®ï¼‰ï¼šä½¿ç”¨è¯¥å·¥å…·è·å–ç”¨æˆ·æ‰€åœ¨çš„ä½ç½®ã€‚

å¦‚æœç”¨æˆ·å‘ä½ è¯¢é—®å¤©æ°”ï¼ŒåŠ¡å¿…ç¡®è®¤æ¸…æ¥šåœ°ç‚¹ã€‚è‹¥èƒ½ä»ç”¨æˆ·çš„é—®é¢˜ä¸­åˆ¤æ–­å‡ºï¼Œä»–ä»¬æƒ³äº†è§£çš„æ˜¯è‡ªå·±å½“å‰æ‰€åœ¨ä½ç½®çš„å¤©æ°”ï¼Œåˆ™ä½¿ç”¨ get_user_location å·¥å…·æŸ¥è¯¢å…¶ä½ç½®ã€‚"""
```

## 2.åˆ›å»ºå·¥å…·
å·¥å…·å…è®¸æ¨¡å‹é€šè¿‡è°ƒç”¨ä½ å®šä¹‰çš„å‡½æ•°ä¸å¤–éƒ¨ç³»ç»Ÿäº¤äº’ã€‚å·¥å…·å¯ä»¥ä¾èµ–è¿è¡Œæ—¶ä¸Šä¸‹æ–‡è¿˜èƒ½ä¸Agent Memoryè¿›è¡Œäº¤äº’ã€‚


```python
from dataclasses import dataclass
from langchain.tools import tool, ToolRuntime

@tool
def get_weather_for_location(city: str) -> str:
    """è·å–ç»™å®šcityçš„å¤©æ°”"""
    return f"{city}çš„å¤©æ°”æ€»æ˜¯æ™´æœ—ï¼"

@dataclass
class Context:
    """ç”¨æˆ·è¿è¡Œæ—¶ä¸Šä¸‹æ–‡çš„æ¨¡å¼"""
    user_id: str
    
@tool
def get_user_location(runtime: ToolRuntime[Context]) -> str:
    """æ ¹æ®ç”¨æˆ·çš„IDæ£€ç´¢ç”¨æˆ·çš„ä¿¡æ¯"""
    user_id = runtime.context.user_id
    
    return "åŒ—äº¬" if user_id == "1" else "ä¸Šæµ·"
```

å·¥å…·åº”å½“æœ‰å®Œå–„çš„æ–‡æ¡£è¯´æ˜ï¼šå®ƒä»¬çš„åç§°ã€æè¿°å’Œå‚æ•°åç§°ä¼šæˆä¸ºæ¨¡å‹æç¤ºè¯çš„ä¸€éƒ¨åˆ†ã€‚LangChain çš„ @tool ä¼šæ·»åŠ å…ƒæ•°æ®ï¼Œå¹¶é€šè¿‡ ToolRuntime å‚æ•°æ”¯æŒè¿è¡Œæ—¶æ³¨å…¥ã€‚

## 3.é…ç½®æ¨¡å‹

```python
from utils import llm_modelscope

model = llm_modelscope(model="Qwen/Qwen3-30B-A3B-Instruct-2507", temperature=0.5, timeout=10, max_tokens=1000)
```

## 4.å®šä¹‰å“åº”responseæ ¼å¼

```python
from dataclasses import dataclass

@dataclass
class ResponseFormat:
    """Agentçš„å“åº”æ¨¡å¼"""
    punny_response: str # ä¸€ä¸ªåŒå…³è¯­çš„å›ç­”ï¼ˆå¿…éœ€çš„ï¼‰
    weather_condition: str | None = None 
```

## 5.æ·»åŠ è®°å¿†
ä¸ºä½ çš„æ™ºèƒ½ä½“æ·»åŠ è®°å¿†ï¼Œä½¿å…¶èƒ½åœ¨äº¤äº’è¿‡ç¨‹ä¸­ä¿æŒçŠ¶æ€ã€‚è¿™èƒ½è®©æ™ºèƒ½ä½“è®°ä½ä¹‹å‰çš„å¯¹è¯å’Œä¸Šä¸‹æ–‡ã€‚  
åœ¨ç”Ÿäº§ç¯å¢ƒä¸­ï¼Œè¯·ä½¿ç”¨å¯æŒä¹…åŒ–çš„æ£€æŸ¥ç‚¹å·¥å…·å°†æ•°æ®ä¿å­˜åˆ°æ•°æ®åº“ã€‚  
https://docs.langchain.com/oss/python/langgraph/add-memory#manage-short-term-memory

```python
from langgraph.checkpoint.memory import InMemorySaver

checkpointer = InMemorySaver()
```

## 6.åˆ›å»ºå¹¶è¿è¡ŒAgent

```python
agent = create_agent(
    model=model,
    tools=[get_weather_for_location, get_user_location],
    system_prompt=SYSTEM_PROMPT,
    response_format=ResponseFormat,
    context_schema=Context,
    checkpointer=checkpointer,
)

# thread_id æ˜¯ä¸€ä¸ªç»™å®šå¯¹è¯çš„å”¯ä¸€æ ‡è¯†ã€‚
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [{"role": "user", "content": "å¤–é¢çš„å¤©æ°”æ€ä¹ˆæ ·ï¼Ÿ"}]},
    config=config,
    context=Context(user_id="1"),
)

print(response["structured_response"])

response = agent.invoke(
    {"messages": [{"role": "user", "content": "å¬æˆ‘è¯´ï¼Œè°¢è°¢ä½ å“¦ï¼"}]},
    config=config,
    context=Context(user_id="1")
)

print(response["structured_response"])
```
